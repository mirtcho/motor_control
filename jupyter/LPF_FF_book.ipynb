```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

l = 1000
#x = np.full((l), 1.0)
y = np.full((l), 0.0)
t = np.arange(0,l)
x=signal.square(2 * np.pi * 20/l * t)

for i in range (2,l):
    y[i]=-0.002*y[i-2]/300-1.6*y[i-1]/300+33*x[i]/300+1.65*x[i-1]/300
fig, axs = plt.subplots(2,figsize=(10,7))
fig.suptitle('Feed forward filter simulation')
axs[0].plot(t,x)
axs[0].set_ylabel('X[t]')
axs[1].plot(t,y)
axs[1].set_ylabel('Y[t]')
plt.show()
```

```python
b = 24
e = 30
fig, axs = plt.subplots(2,figsize=(12,7))
fig.suptitle('Zoomed in')
axs[0].plot(t[b:e],x[b:e])
axs[0].set_ylabel('x[t]')
axs[1].plot(t[b:e],y[b:e])
axs[1].set_ylabel('y[t]')
plt.show()
print (y[22:29])
```
