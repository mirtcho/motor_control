```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

x = np.arange(-2,6,0.001)
y = 1/2 * x**2     - 2*x + 6
y1= 1/2 * x**2     - 2*x + 5
y2= 1/2 * (x+1)**2 - 2*(x+1) + 6

#plots
fig, axs = plt.subplots(2,figsize=(10,17))
fig.suptitle('Vertically stacked subplots')
plt.ylabel('some numbers')
axs[0].plot(x,y,x,y1)
axs[1].plot(x,y,x,y2)
plt.show()
```

```python
# EXP snd pt1 function
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
#
step=0.01
t = np.arange(-0.1,6,step)
x = np.zeros(t.size)
y1= np.zeros(t.size)
y = 1-np.exp(-t)
y[:10]=0
x[10:]=1
ind=0
for i in x:    
    if ind>1:
       y1[ind] = y1[ind-1] + 4*step*(x[ind]-y1[ind-1])
    ind = ind + 1

#plots
plt.rcParams["figure.figsize"] = (16,10)
plt.xlabel('X[m]')
plt.ylabel('H[m]')
plt.plot(t,x,t,y,t,y1)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import math as m

x = np.linspace(0,2*m.pi,2000)
Ua = []
Ub = []
Uc = []
A = []
ind=0
for t in x:
  Ua.append (1* m.cos(t))
  Ub.append (1 * m.cos(t+2*m.pi/3))
  Uc.append (1 * m.cos(t+4*m.pi/3))
  Alpha   = 2*Ua[ind]/m.sqrt(3) - Ub[ind]/3 - Uc[ind]/3
  Beta    = Ub[ind]/m.sqrt(3)   - Uc[ind]
  A.append (m.sqrt(Alpha*Alpha+Beta*Beta))
  ind = ind + 1

Va = []
Vb = []
Vc = []
An = []
ind=0
for t in x:
  Va.append (1* m.cos(t))
  Vc.append (1 * m.cos(t+2*m.pi/3))
  Vb.append (1 * m.cos(t+4*m.pi/3))
  Alpha   = 2*Ua[ind]/m.sqrt(3) - Ub[ind]/3 - Uc[ind]/3
  Beta    = Ub[ind]/m.sqrt(3)   - Uc[ind]
  An.append (m.sqrt(Alpha*Alpha+Beta*Beta))
  ind = ind + 1


#plots
fig, axs = plt.subplots(2,figsize=(10,17))
fig.suptitle('Vertically stacked subplots')
plt.ylabel('some numbers')
axs[0].plot(x,Ua,x,Ub,x,Uc)
axs[1].plot(x,A,x,An)
plt.show()
```
