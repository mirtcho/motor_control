```python
import matplotlib.pyplot as plot
import numpy as np
import math as m

class Test():
  def __init__(self):
    self.create_inputs()

  def test1(self):
    #self.inputs_alpha_beta()
    self.delta_eye_cor()
    
  def create_inputs(self):
    self.time = np.arange(0, 4*m.pi, 0.01);
    # Amplitude of the reference signals. They are artificialy created
    self.u_l1n_ref = np.sin(self.time)
    self.u_l2n_ref = np.sin(self.time  + 2*m.pi/3.0)
    self.u_l3n_ref = np.sin(self.time  - 2*m.pi/3.0)
    #create Bidigi signals from reference signals
    self.u_l1n = self.u_l1n_ref
    self.u_l12  = self.u_l2n_ref - self.u_l1n_ref
    self.u_l23  = self.u_l3n_ref - self.u_l2n_ref
    self.u_l31  = self.u_l1n_ref - self.u_l3n_ref
    
  def delta_eye_cor(self):
    # create line to N from Bidig data
    # made only for alpha
    self.u_l1_cor = self.u_l1n
    self.u_l2_cor = self.u_l12 + self.u_l1n 
    self.u_l3_cor = self.u_l31 - self.u_l1n
    
  def plot_ref(self):
    # Plot a sine wave using time and amplitude obtained for the sine wave
   plot.plot(self.time, self.u_l1n_ref, 
             self.time, self.u_l2n_ref,
             self.time, self.u_l3n_ref)   
   plot.title('The reference Ul1,2,3 to N')
   # Give x axis label for the sine wave plot
   plot.xlabel('Time')
   # Give y axis label for the sine wave plot
   plot.ylabel('Ref Amplitude = 1')
   plot.grid(True, which='both')
   plot.axhline(y=0, color='k')
   plot.show()

  def plot_bidi(self):   
   fig, ax = plot.subplots()
   ax.plot(self.time, self.u_l12, '-r', label='12')
   ax.plot(self.time, self.u_l23, '-g', label='23')
   ax.plot(self.time, self.u_l31, '-b', label='31')
   plot.title('Bidi measurements')
   # Give x axis label for the sine wave plot
   plot.xlabel('Time')
   # Give y axis label for the sine wave plot
   plot.ylabel('Ref Amplitude = x')
   plot.legend(loc='upper left', frameon=False)
   plot.grid(True, which='both')
   plot.axhline(y=0, color='k')
   plot.show()

  def plot_cor(self):
   self.delta_eye_cor()
   fig, ax = plot.subplots()
   ax.plot(self.time, self.u_l1_cor, '-r', label='1')
   ax.plot(self.time, self.u_l2_cor, '-g', label='2')
   ax.plot(self.time, self.u_l3_cor, '-b', label='3')
   plot.title('Alpha Ul1,2,3 to N caclc Cor')
   # Give x axis label for the sine wave plot
   plot.xlabel('Time')
   # Give y axis label for the sine wave plot
   plot.ylabel('Ref Amplitude = x')
   plot.legend(loc='upper left', frameon=False)
   plot.grid(True, which='both')
   plot.axhline(y=0, color='k')
   plot.show()
```

```python
t = Test()
t.test1()
t.plot_ref()
```
